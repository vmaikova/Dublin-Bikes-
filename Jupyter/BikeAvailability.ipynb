```python
def getWeatherDetailsForDay(day):
    jsonData = pa.read_json('wunderground/Dublin04' + str(day) + '.json')
    #display(jsonData['history']['observations'])
    #display(type(jsonData))
    dictionary = {}
    for date in jsonData['history']['observations']:
        dictionary[date['date']['hour'] + ':' + date['date']['min']] = [date['rain'], date['tempm'], date['conds']]
        #display(date['date']['hour'] + ' ' + date['date']['min'] + ' ' + date['rain'] + ' ' + date['tempm'] + ' ' + date['conds'])
    return dictionary
days = getWeatherDetailsForDay(16)
```

```python
def roundToNearestHalfHourAndRemoveDate(dateTime):
    date = parse(dateTime)
    if(date.minute > 45):
        if(date.hour + 1 == 24):
            return str(date.hour) + ":" + str(30)
        return str(date.hour + 1) + ":" + "00"
    if(date.minute > 15):
        return str(date.hour) + ":" + str(30)
    return str(date.hour) + ":" + "00"
```

```python
df = df.sample(frac=1).reset_index(drop=True)
df = df[:30000]
X = df[[
    'Number',
    'Last_update',
    'Available_bike_stands'
]]
y = df[
    'Available_bikes'
]
y2 = df[
    'Available_bike_stands'
]

len(X)
```

```python
X['Rain'] = ''
X['Temp'] = ''
X['Description'] = ''
X['Time'] = ''

for i in range(0, len(X)):
    display(i)
    time = roundToNearestHalfHourAndRemoveDate(X.iloc[i]['Last_update'])
    if(len(time) == 4):
        time = "0" + time
    X.at[i, 'Time'] = time
    weatherData = getWeatherDetailsForDay(parse(X.iloc[i]['Last_update']).day)
    X.at[i, 'Rain'] = weatherData[time][0]
    X.at[i, 'Temp'] = float(weatherData[time][1])
    X.at[i, 'Description'] = weatherData[time][2]
```

```python
X = X[[
    'Rain',
    'Temp',
    'Description',
    'Time',
    'Available_bike_stands'
]]
```

```python
X['Rain'] = pa.Categorical(X['Rain'], categories=["0", "1"]) #.astype('category')
X['Description'] = pa.Categorical(X['Description'], categories=['Mostly Cloudy', 'Light Drizzle', 'Light Rain', 'Overcast','Partly Cloudy', 'Fog', 'Clear', 'Scattered Clouds', ''])
X['Time'] = pa.Categorical(X['Time'], categories=['19:30', '20:00', '20:30', '21:00', '21:30', '22:00', '22:30','23:00', '23:30', '00:00', '00:30', '01:00', '01:30', '02:00','02:30', '03:00', '03:30', '04:00', '04:30', '05:00', '05:30','06:00', '06:30', '07:00', '07:30', '08:00', '08:30', '09:00', '09:30', '10:00', '10:30', '11:00', '11:30', '12:00', '12:30', '13:00', '13:30', '14:00', '14:30', '15:00', '15:30', '16:00', '16:30', '17:00', '17:30', '18:00', '18:30', '19:00'])
X['Temp'] = pa.to_numeric(X['Temp'])
X['Available_bike_stands'] = pa.to_numeric(X['Available_bike_stands'])
#pa.to_numeric(X['Temp'])
```

```python
X = pa.concat([X, y], axis=1, join_axes=[X.index])
X = X.dropna(subset = ['Rain', 'Description', 'Time', 'Temp', 'Available_bike_stands'])
y = X['Available_bikes']
```

```python
categorical = X[['Description', 'Time', 'Rain']].apply(LabelEncoder().fit_transform)
```

```python
continous = X[['Temp', 'Available_bike_stands']]
```

```python
clean = pa.concat([categorical, continous], axis=1, join_axes=[categorical.index])
```

```python
def createWeatherModelForStation( dataset, target):
    #station_df = dataset.loc[df['Number'] == stationNumber]
    rfc = RandomForestClassifier()
    rfc.fit(dataset, target)
    return rfc
```

```python
model = createWeatherModelForStation(clean, y)
```
